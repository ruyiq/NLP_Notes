```python
#regular expressions allow for pattern searching in a text document
# \d (a special code)
# r'\d{3}-\d{3}-\d{4}'
```

```python
text = "The phone number of the agent is 408-555-1034. Call soon!"
```

```python
"phone" in text
```

```python
import re #regular expression library in python
```

```python
pattern = "phone"
```

```python
re.search(pattern, text)
```

```python
my_match = re.search(pattern, text)
```

```python
my_match.span() #from index 4 of the string and goes up to 9
```

```python
my_match.end()
```

```python
text = "my phone is a new phone"
```

```python
match = re.search(pattern, text)
```

```python
match.span() #can only find the first; try findall()
```

```python
all_matches = re.findall("phone", text)
```

```python
all_matches
```

```python
len(all_matches)
```

```python
for match in re.finditer("phone", text):
    print(match.span())
```

```python
phrase= 'There are 5 numbers in the string: 34, 56, 65, 33, 21'
```

```python
re.findall(r"[^\d]", phrase) #"" means exclude
```

```python
re.findall(r"[^\d]+", phrase)
```

```python
test_phrase = "This is a string ! but it has punctuation. How to remove it?"
```
